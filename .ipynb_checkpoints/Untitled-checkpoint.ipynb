```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
players = pd.read_csv('players.csv', index_col = 0)
```

```python
players.head()
```

```python
convert_to_binary = ['first_blood_assist', 
                     'first_blood_kill', 
                     'first_tower_assist', 
                     'first_tower_kill', 
                     'first_inhibitor_assist', 
                     'first_inhibitor_kill',
                     'victory']

for column in convert_to_binary:
    players[column] = players[column].map({True: 1, False: 0})
```
